```python
from bs4 import BeautifulSoup
import requests

soup = BeautifulSoup(open("List of postal codes of Canada_ M - Wikipedia.htm"))
FSAs_table = soup.table # get table part from the website 
FSAs_table
```

```python
FSAs_record_list=[]
for record in FSAs_table.find_all('tr'):
    FSAs_record_ori = record.text.split('\n')
    FSAs_record = list(filter(None, FSAs_record_ori)) # delete empty element 
    if FSAs_record[1] != 'Not assigned': # delete not assgined record
        FSAs_record[2] = FSAs_record[2].replace(' /',',') # replace / with , in neighbohood column 
        FSAs_record_list.append(FSAs_record)
del FSAs_record_list[0] # delete the first title line 
FSAs_record_list
```

```python
import pandas
from pandas import DataFrame
df=DataFrame (FSAs_record_list,columns=['Postal Code', 'Borough', 'Neighborhood'])
df
```
